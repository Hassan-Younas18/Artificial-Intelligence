```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Extract the variables for the scatter plot
sleep_hours = df['Sleep Hours']
productivity_rank = df['Productivity Rank']

# Create the scatter plot
plt.figure(figsize=(6, 4))
plt.scatter(sleep_hours, productivity_rank, color='teal', marker='o', alpha=0.7, edgecolor='black')
plt.title('Sleep Hours vs. Productivity Rank', fontsize=12)
plt.xlabel('Sleep Hours', fontsize=14)
plt.xticks(fontsize=12)
plt.xlim(0, 14)
plt.ylabel('Productivity Rank', fontsize=14)
plt.yticks(fontsize=12)
plt.ylim(0, 14)
plt.grid(False)

# Fit a linear regression model
model = LinearRegression()
model.fit(sleep_hours.values.reshape(-1, 1), productivity_rank)

# Get the coefficients
slope = model.coef_[0]
intercept = model.intercept_


variable1 = 'Sleep Hours'
variable2 = 'Productivity Rank'

# Calculate the correlation coefficient between the two variables
correlation = df[variable1].corr(df[variable2])

# Display the correlation coefficient
print(f"Correlation between '{variable1}' and '{variable2}': {correlation}")

# Plot the regression line
x_values = np.linspace(0, 14, 100)
y_values = slope * x_values + intercept
plt.plot(x_values, y_values, color='green', linestyle='-', linewidth=2, label=f'Line of Best Fit (Correlation: {correlation:.2f})')
plt.legend()

plt.tight_layout()
plt.show()

```

# Implementing Gradient Descent

```python
import numpy as np

# Define the dataset
# Assuming the dataset is stored in a variable named 'routine_dataset'
# Each row corresponds to a day, with columns for independent variables and the target variable
X = df[['Sleep Hours', 'Food Rank', 'Working Hours', 'Health Rank', 'Mood Rank']]
y = df['Productivity Rank']

# Number of observations
m = len(y)

# Initialize coefficients and intercept randomly
np.random.seed(0)  # For reproducibility
m_coefficients = np.random.rand(X.shape[1])
c_intercept = np.random.rand()

# Hyperparameters
learning_rate = 0.001
no_of_epochs = 500

# Gradient Descent
for _ in range(no_of_epochs):
    # Calculate predicted values
    y_pred = np.dot(X, m_coefficients) + c_intercept
    
    # Calculate error
    error = y_pred - y
    
    # Update coefficients and intercept
    m_coefficients -= (2/m) * learning_rate * np.dot(X.T, error)
    c_intercept -= (2/m) * learning_rate * np.sum(error)

# Print final coefficients and intercept
print("Coefficients:", m_coefficients)
print("Intercept:", c_intercept)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming 'routine_dataset' contains the dataset
X = df[['Sleep Hours', 'Food Rank', 'Working Hours', 'Health Rank', 'Mood Rank']]
y = df['Productivity Rank']

# Number of observations
m = len(y)

# Initialize coefficients and intercept randomly
np.random.seed(0)  # For reproducibility
m_coefficients = np.random.rand(X.shape[1])
c_intercept = np.random.rand()

# Hyperparameters
learning_rate = 0.001
no_of_epochs = 1000

# Track the changes in coefficients, intercept, and loss
coefficients_history = []
intercept_history = []
losses = []

# Gradient Descent
for epoch in range(no_of_epochs):
    # Calculate predicted values
    y_pred = np.dot(X, m_coefficients) + c_intercept
    
    # Calculate error
    error = y_pred - y
    
    # Update coefficients and intercept
    m_coefficients -= (2/m) * learning_rate * np.dot(X.T, error)
    c_intercept -= (2/m) * learning_rate * np.sum(error)

    # Calculate loss
    loss = np.mean(error ** 2)
    
    # Track history
    coefficients_history.append(m_coefficients.copy())
    intercept_history.append(c_intercept)
    losses.append(loss)

# Plot the working of gradient descent
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
for i in range(X.shape[1]):
    plt.plot(range(no_of_epochs), [coeffs[i] for coeffs in coefficients_history], label=f'Coefficient {i+1}')
plt.xlabel('Epoch')
plt.ylabel('Coefficient Value')
plt.title('Changes in Coefficients over Epochs')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(range(no_of_epochs), intercept_history, label='Intercept')
plt.xlabel('Epoch')
plt.ylabel('Intercept Value')
plt.title('Changes in Intercept over Epochs')
plt.legend()
plt.show()
```
